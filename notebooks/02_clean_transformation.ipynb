## Creación de variables derivadas

Con el fin de enriquecer el análisis y facilitar la identificación de patrones, se crearon dos variables adicionales a partir de los indicadores originales del conjunto de datos.


### Índice de Riesgo Académico

Esta variable corresponde a la suma de los porcentajes de deserción, reprobación y repitencia.

```python
indice_riesgo = desercion + reprobacion + repitencia
```

Este indicador se utiliza como un **índice compuesto** para identificar departamentos con mayor concentración de factores asociados al riesgo académico. No representa un porcentaje real de estudiantes, sino una métrica de apoyo para comparar el nivel de riesgo entre departamentos.


````python
df['riesgo_academico'] = (
    df['desercion'] +
    df['reprobacion'] +
    df['repitencia']
)
````

### Balance Académico

Esta variable se calcula como 
```python
balance_academico = aprobacion - reprobacion
```

Su objetivo es medir el desempeño académico general de cada departamento. Valores más altos indican un mejor rendimiento, mientras que valores bajos pueden sugerir mayores dificultades académicas.
la diferencia entre el porcentaje de aprobación y el porcentaje de reprobación.

````python
df['balance_academico'] = (
    df['aprobacion'] - 
    df['reprobacion']
)
````
